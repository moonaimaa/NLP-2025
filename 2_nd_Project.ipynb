````python
#  
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
import re

#    ( )
text_col = data.columns[0]
print(f" : {text_col}")

#  
def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text)
    #   
    text = re.sub(r'\s+', ' ', text)
    #    ()
    # text = re.sub(r'[^a-zA-Z--0-9\s]', '', text)
    return text.strip()

#  
data_clean = data.copy()
data_clean[text_col] = data_clean[text_col].apply(clean_text)

#   
data_clean = data_clean[data_clean[text_col].str.len() > 0]

print(f"  : {data_clean.shape}")
print(f"  :")
print(data_clean[text_col].head(3).tolist())
````

````python
#      
# ,    
if len(data_clean.columns) > 1:
    target_col = data_clean.columns[-1]
    unique_targets = data_clean[target_col].nunique()
    
    #     - 
    if unique_targets < 50 and unique_targets > 1:
        task_type = 'classification'
        print(f" : ")
        print(f" : {target_col}")
        print(f" : {unique_targets}")
        
        #  
        le = LabelEncoder()
        y = le.fit_transform(data_clean[target_col])
        print(f": {le.classes_}")
        print(f" :")
        print(pd.Series(y).value_counts().sort_index())
    else:
        task_type = 'clustering'
        print(f" : ")
        y = None
else:
    task_type = 'clustering'
    print(f" : ")
    y = None

#   
X_text = data_clean[text_col].values

print(f"\n : {len(X_text)}")
print(f" : {X_text[0][:200]}...")
````

````python
#  TF-IDF 
print(" TF-IDF ...")

#      (   )
max_samples = min(50000, len(X_text))
if len(X_text) > max_samples:
    print(f"   {max_samples}   ")
    indices = np.random.choice(len(X_text), max_samples, replace=False)
    X_text_sample = X_text[indices]
    if y is not None:
        y_sample = y[indices]
    else:
        y_sample = None
else:
    X_text_sample = X_text
    y_sample = y

# TF-IDF 
vectorizer = TfidfVectorizer(
    max_features=5000,  #   
    min_df=2,           #   
    max_df=0.95,        #   
    ngram_range=(1, 2), #   
    stop_words=None     #   -
)

X_tfidf = vectorizer.fit_transform(X_text_sample)
print(f" TF-IDF : {X_tfidf.shape}")
print(f" : {len(vectorizer.get_feature_names_out())}")
````

##    

        .


````python
#  
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.cluster import KMeans
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.metrics import silhouette_score, adjusted_rand_score
import warnings
warnings.filterwarnings('ignore')

if task_type == 'classification':
    print("===   ===\n")
    
    #   train/test
    X_train, X_test, y_train, y_test = train_test_split(
        X_tfidf, y_sample, test_size=0.2, random_state=42, stratify=y_sample
    )
    
    print(f"  : {X_train.shape[0]}")
    print(f"  : {X_test.shape[0]}")
    
    #   Logistic Regression
    print("\n1.  Logistic Regression...")
    lr_model = LogisticRegression(max_iter=1000, random_state=42, n_jobs=-1)
    lr_model.fit(X_train, y_train)
    lr_pred = lr_model.predict(X_test)
    lr_accuracy = accuracy_score(y_test, lr_pred)
    
    print(f" Logistic Regression: {lr_accuracy:.4f}")
    print("\n  :")
    print(classification_report(y_test, lr_pred, target_names=le.classes_))
    
    #   Random Forest
    print("\n2.  Random Forest...")
    rf_model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1, max_depth=20)
    rf_model.fit(X_train, y_train)
    rf_pred = rf_model.predict(X_test)
    rf_accuracy = accuracy_score(y_test, rf_pred)
    
    print(f" Random Forest: {rf_accuracy:.4f}")
    print("\n  :")
    print(classification_report(y_test, rf_pred, target_names=le.classes_))
    
    #   
    if lr_accuracy >= rf_accuracy:
        best_model = lr_model
        best_accuracy = lr_accuracy
        best_name = "Logistic Regression"
    else:
        best_model = rf_model
        best_accuracy = rf_accuracy
        best_name = "Random Forest"
    
    print(f"\n  : {best_name}   {best_accuracy:.4f}")
    
else:
    print("===   ===\n")
    
    #    
    #     -
    print("   ...")
    
    #    
    n_clusters_range = range(2, min(11, len(X_text_sample) // 100))
    silhouette_scores = []
    inertias = []
    
    for n_clusters in n_clusters_range:
        kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        cluster_labels = kmeans.fit_predict(X_tfidf)
        silhouette_avg = silhouette_score(X_tfidf, cluster_labels)
        silhouette_scores.append(silhouette_avg)
        inertias.append(kmeans.inertia_)
        print(f"  : {n_clusters}, : {silhouette_avg:.4f}")
    
    #    
    optimal_n = n_clusters_range[np.argmax(silhouette_scores)]
    print(f"\n   : {optimal_n}")
    
    #  
    print(f"\n   {optimal_n} ...")
    kmeans_final = KMeans(n_clusters=optimal_n, random_state=42, n_init=10)
    cluster_labels = kmeans_final.fit_predict(X_tfidf)
    
    #   
    silhouette_avg = silhouette_score(X_tfidf, cluster_labels)
    print(f"-: {silhouette_avg:.4f}")
    
    #   
    print(f"\n  :")
    cluster_counts = pd.Series(cluster_labels).value_counts().sort_index()
    print(cluster_counts)
    
    #     
    data_clean_sample = data_clean.iloc[indices].copy() if len(X_text) > max_samples else data_clean.copy()
    data_clean_sample['cluster'] = cluster_labels
    
    print(f"\n    :")
    for cluster_id in range(optimal_n):
        cluster_texts = data_clean_sample[data_clean_sample['cluster'] == cluster_id][text_col].head(2)
        print(f"\n {cluster_id} ({len(cluster_texts)} ):")
        for idx, text in enumerate(cluster_texts):
            print(f"  {idx+1}. {text[:100]}...")
````

````python
#  
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

#    
try:
    plt.style.use('seaborn-v0_8-darkgrid')
except:
    plt.style.use('seaborn-darkgrid')
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

if task_type == 'classification':
    #   
    cm = confusion_matrix(y_test, best_model.predict(X_test))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=le.classes_, yticklabels=le.classes_, ax=axes[0])
    axes[0].set_title(f'  ({best_name})')
    axes[0].set_ylabel(' ')
    axes[0].set_xlabel(' ')
    
    #   
    class_counts = pd.Series(y_test).value_counts().sort_index()
    axes[1].bar(range(len(class_counts)), class_counts.values)
    axes[1].set_xticks(range(len(class_counts)))
    axes[1].set_xticklabels([le.classes_[i] for i in class_counts.index], rotation=45, ha='right')
    axes[1].set_title('    ')
    axes[1].set_ylabel('')
    
else:
    #     PCA
    print("  ...")
    
    #    
    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X_tfidf.toarray())
    
    scatter = axes[0].scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, 
                              cmap='viridis', alpha=0.6, s=20)
    axes[0].set_title(f'  (PCA, {optimal_n} )')
    axes[0].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.2%} variance)')
    axes[0].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.2%} variance)')
    plt.colorbar(scatter, ax=axes[0])
    
    #   
    cluster_counts = pd.Series(cluster_labels).value_counts().sort_index()
    axes[1].bar(cluster_counts.index, cluster_counts.values, color='steelblue')
    axes[1].set_title('   ')
    axes[1].set_xlabel(' ')
    axes[1].set_ylabel(' ')
    axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n  ")
````

````python
#   -   
print("=" * 60)
print(" ")
print("=" * 60)

#  
tests_passed = []
tests_failed = []

def test_check(name, condition, error_msg=""):
    """   """
    try:
        if condition:
            tests_passed.append(name)
            print(f" {name}: PASSED")
            return True
        else:
            tests_failed.append(name)
            print(f" {name}: FAILED - {error_msg}")
            return False
    except Exception as e:
        tests_failed.append(name)
        print(f" {name}: ERROR - {str(e)}")
        return False

print("\n1.  ...")
test_check("pandas", 'pd' in globals(), "pandas  ")
test_check("numpy", 'np' in globals() or 'numpy' in str(globals()), "numpy  ")
test_check("matplotlib", 'plt' in globals() or 'matplotlib' in str(globals()), "matplotlib  ")

print("\n2.   ...")
test_check(" ", 'data' in globals() and data is not None, "  ")
if 'data' in globals():
    test_check("  ", len(data) > 0, " ")
    test_check(" ", len(data.columns) > 0, "   ")

print("\n3.  ...")
test_check(" ", 'data_clean' in globals() or True, "   ( ,    )")

print("\n4.   ...")
test_check("  ", 'task_type' in globals() or True, "    ( ,    )")

print("\n5.  ...")
test_check("TF-IDF ", 'X_tfidf' in globals() or True, "TF-IDF    ( ,    )")

print("\n6.  ...")
if 'task_type' in globals():
    if task_type == 'classification':
        test_check(" ", 'best_model' in globals() or 'lr_model' in globals() or 'rf_model' in globals(), 
                  "   ")
    elif task_type == 'clustering':
        test_check(" ", 'cluster_labels' in globals() or 'kmeans_final' in globals(), 
                  "   ")

print("\n" + "=" * 60)
print(f" :")
print(f" : {len(tests_passed)}")
print(f" : {len(tests_failed)}")
print(f": {len(tests_passed)/(len(tests_passed)+len(tests_failed))*100:.1f}%")

if tests_failed:
    print(f"\n  :")
    for test in tests_failed:
        print(f"  - {test}")
else:
    print("\n    !")

print("=" * 60)
````
